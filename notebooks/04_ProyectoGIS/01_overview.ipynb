# GIS Project

This unit introduces practical work with geographic information published through web services. The goal is to learn how to consume, visualize, and combine geospatial data with Python, starting from remote information and producing simple cartographic outputs and vector analyses.

The estimated time to complete the notebooks and their activities is **4 hours**.

## Notebooks in this unit

### 02) WMS and visualization

This notebook works with [**WMS** services](https://es.wikipedia.org/wiki/Web_Map_Service) and geographic visualization techniques. It uses several Python libraries to request maps, manipulate raw responses, interpret service parameters, and build useful visualizations for exploring raster layers or images generated by map servers.

The focus is to understand what a WMS returns, how requests are built, and how that information can be integrated into a reproducible Python workflow.

We'll use the following libraries:
- OWSLib. A WMS client 
- Rasterio. A library for working with georeferenced images (GeoTIFFs) and operations such as transforms and CRS.


### 03) WFS, GeoJSON, vector data, and GeoPandas

This notebook introduces vector data consumption from **WFS** services and APIs that publish **GeoJSON**. The workflow uses `geopandas` `GeoDataFrame` objects, attribute filtering, coordinate reference systems, reprojection, basic spatial operations, and layer combination in a single map.

The main example combines public schools and the road network to show how to integrate two different geographic sources in one visualization and obtain simple metrics.
