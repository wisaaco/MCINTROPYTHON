# Proyecto GIS

En este bloque se plantea una introduccion practica al trabajo con informacion geografica publicada en servicios web. El objetivo es aprender a consumir, visualizar y combinar datos geoespaciales usando Python, partiendo de informacion remota y llegando a productos cartograficos y analisis vectoriales sencillos.

La duracion estimada para realizar los notebooks y sus actividades es de **3 horas**.

## Notebooks del bloque

### 02) WMS y visualizacion

Este notebook trabaja con servicios **WMS** y con tecnicas de visualizacion de informacion geografica. Se plantea el uso de varias librerias de Python para solicitar mapas, manipular respuestas en crudo, interpretar parametros del servicio y construir visualizaciones utiles para explorar capas raster o imagenes generadas por servidores cartograficos.

El enfoque esta pensado para entender que devuelve un WMS, como se construyen las peticiones y como se puede integrar esa informacion en un flujo de trabajo reproducible con Python.

### 03) WFS, GeoJSON, datos vectoriales y GeoPandas

Este notebook introduce el consumo de datos vectoriales desde servicios **WFS** y desde APIs que publican **GeoJSON**. A partir de esos datos se trabaja con `GeoDataFrame` de `geopandas`, filtrado de atributos, sistemas de referencia, reproyecciones, operaciones espaciales basicas y combinacion de capas en un mismo mapa.

El ejemplo principal combina colegios publicos y red de carreteras para mostrar como integrar dos fuentes geograficas distintas en una misma visualizacion y obtener metricas sencillas.

## Actividades

Los notebooks incluyen actividades propuestas para practicar lo visto: modificar parametros de servicios web, probar otras capas, filtrar datos por criterios espaciales o atributivos, reproyectar informacion y comparar formatos de datos geograficos.
