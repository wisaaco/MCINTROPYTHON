## Session 1.2: Program Flow Control

A program usually consists of more than one instruction. More complex programs need to execute a series of statements repeatedly or choose between alternative paths based on the value of a variable at runtime.

To implement these ideas, we need to understand the different control structures that programming languages provide. A control structure directs the execution order of a program's instructions. All control structures have a single entry point and a single exit point. This is one of the principles behind structured programming. Structured programs are built from simple, hierarchically organized structures that control the flow of execution.

### Conditional

In the real world, there are many situations where we must evaluate the available information and then choose between several options based on what we observed. These patterns cannot be reproduced with simple sequential composition, so we introduce conditional composition.

This pattern corresponds to structures like:
```
if it rains
   then I take an umbrella
else
   then I take a hat
```

In a Python program, the `if` statement is used to make this kind of decision. It allows conditional execution of a statement or group of statements based on the value of an expression. Its structure is:
```
if <expr>:
    <body>
```

In the code above, `<expr>` is an expression evaluated in a boolean context and `<body>` is one or more valid Python expressions. Also note that the body must be indented.

In the following example we use conditional composition to obtain the absolute value of a number:

````python
# Program that calculates the absolute value of a number
x = 12

# Conditional block
if x < 0:
    print("Entering the conditional block")
    x = x * -1  # x *= -1 or x = -x

print("We are outside the conditional block")
# Results
print(f'The absolute value is: {x}')
````

#### The `else` and `elif` Clauses

Now that we know how to use an `if` statement to execute a single statement or a block of statements conditionally, it is time to see what else we can do.

Some problems require evaluating a condition and taking one path if it is true, but also taking an alternative path if it is not. This is achieved with an `else` clause. For example, if we want a program that tells us whether a number is positive or negative, we can write it like this:

````python
x = -1

if x > 0:
    print("Evaluating the number")  # A block with two code statements
    print("The number is positive")
else:
    print("The number is negative")
````

The previous code does not handle the case when `x` is 0. To complete and correct our program, we need to evaluate one more condition. We can do this with the `elif` keyword.

````python
x = 0

if x > 0:
     print("Evaluating the number") # A block with two code statements.
     print("The number is positive")

elif x == 0: # Here we handle the case when the number is 0

     print("The number is zero")

else:
# Here we do not need another condition. If the number is not positive
# or zero, it must be negative

     print("The number is negative")

print("See you next time")
````

In general, we can have as many conditions as we want or need:
```
if <expr>:
    <statements>
elif <expr>:
    <statements>
elif <expr>:
    <statements>
    ...
else:
    <statements>
```

### Loops

Other problems require repeating the same block of code more than once. Think, for example, about the code needed to calculate a multiplication table: it basically consists of repeating the same expression while changing one of the operands on each iteration. The control structure that implements iterative composition is called a loop.

In programming, there are two types of iteration: undefined and defined:

   - **Undefined iteration**: the number of loop repetitions is not explicit. Instead, the designated block is executed repeatedly as long as some logical condition is true, that is, a _boolean_ condition.

   - **Defined iteration**: the number of times the block will execute is explicit when the loop starts. This can be done using either an integer variable or a literal.

#### The indefinite loop

To create iterations with an initially undefined number of repetitions, we use the `while` statement. In this section we will see how this statement is used to build programs with loops.

The format of a `while` loop is shown below:

```
while <expr>:
    <body>
```

When the loop is encountered, the expression `<expr>` is evaluated once in a boolean context. If the result is true, the body of the loop is executed. Then `<expr>` is evaluated again, and if it is still true, the body executes again. This process continues until `<expr>` evaluates to `False`, at which point execution continues with the first statement after the loop body. Notice that the loop body is indicated by indentation under the `while` statement.

An example of a program that uses an indefinite loop is the following. This program performs a countdown starting from the value of the variable `n` down to 1:

````python
n = 5
print("The initial value of variable 'n' is: " + str(5))
while n > 0: # What happens when n = 0?

    print(n)
    n = n - 1

print(f'Value of n = {n}')
print("Out!")
````

#### The defined loop

This type of loop is more controlled than the previous one, because it does not end when a condition stops being true, but rather because we know how many iterations it will perform from the beginning. The statement used in this case is called `for`, and we explain it below.

```
for <var> in <iterable>:
     <statement(s)>
```

`var` is the variable used within the loop body and does not need to be defined beforehand. `<iterable>` is a
collection of objects to iterate over, for example a list or tuple. `<statement(s)>` in the loop body is denoted by indentation as explained for conditionals and while loops; all expressions in `<statement(s)>` execute once for each element of `<iterable>`. The loop variable `<var>` takes the value of each element in `<iterable>`. This type of loop is often used with data structures such as lists or tuples, which we will see in the following section.

Below are two illustrative examples:


````python
for i in (0,1,2,3,4,5):
    s = i + 3
    print(s)
````

````python
for i in range(10):
    print(i)
````

The function `range(<start>, <end>, <step>)` returns an iterable that produces integers starting at `<start>` and ending at `<end> - 1`. If specified, `<step>` indicates the increment between consecutive values in the sequence.
