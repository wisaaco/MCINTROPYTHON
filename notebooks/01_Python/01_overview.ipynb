### Type

The type defines the nature of the values that a variable can have. It also limits the operations that can be performed with them.

The elementary data types are: character, integer, real, and boolean:

**Character** allows you to represent information about simple characters and sequences of characters, which are represented between quotes, either single \' or double \". In the world of programming, we know a sequence of characters by the name of _String_ (_str_). Python considers characters as _strings_ of length 1.

- Examples of characters: 'a', 'b', '\n'.
- Examples of _String_: '3', "hello", "John".

**Integer** (*int*) allows you to represent numerical values, both positive and negative.

- Examples of integers: -33, 14, 28, or 8545445.

**Real** (*float*): this type designates a floating-point number. These values are specified with a decimal point. Optionally, you can add the character _e_ or _E_ followed by a positive or negative integer to specify the number in scientific notation.

- Examples of real numbers: -1.2, 3.1416, -0.26989, 4. , 3e7 (3000000.0), 1.6e-4 (0.00016).

**Complex**: numerical values in complex format. They are specified as <real part>+<imaginary part>j.
- Examples of complex numbers: 2+3j, 33+0j.

**Boolean**: logical values. They can have only two values: true (`True`) or false (`False`).

The maximum and minimum values that we can store in an integer or real type variable are not defined, unlike in other languages, they depend on the machine where the code is executed.

### Instructions, Operators, and Operands

An **instruction** is a command that tells the program what to do. It's an action that the program executes, often using variables to perform calculations or produce results. In Python, **operators** are special symbols that indicate that some type of computation is necessary, and the values on which an operator acts are called **operands**.

**Characteristics of Instructions:**
- Performs an **action**: e.g., print, calculate, store data
- Often **operates on variables**: uses variable values to accomplish tasks. These instructions are also known as **operators**.
- Executed in **order**: from top to bottom (unless control structures change the flow)
- Can produce **output**: results that affect the program's behavior

**Examples of Instructions:**

```python
print(age)                          # Instruction: print the variable 'age'
result = age + 5                    # Instruction: add 5 to age and store in 'result'
print(f"Next year I'll be {age + 1}") # Instruction: print a message with a calculation
```

**Understanding Operators and Operands:**
We will understand operators and operands with a small example:


````python
x = 10.0
y = 5
result = x + y
````

In this case, the `+` operator allows us to add the operands `x` and `y` together and save the result in the variable of the same name. We must highlight that in this code we have needed to define three different variables.

An operand can be a variable or literal. We define **literals** in a program as the idea of expressing a value that does not change in the source code, these must be of one of the different data types that we have described previously.

Here is another code snippet using other operators and integer type literals:

````python
a = 10
b = 20
x = 45
r = (a + b - 5) + (x + 10 + 20)
````

**How They Work Together:**

Variables store the **data**, and instructions **manipulate** that data using operators:

````python
# Variables store data
price = 10.50
quantity = 3

# Instructions work with variables using operators
print(f"Total: ${price * quantity}")          # Display the result
````


In this example:
- `price` and `quantity` are **variables** holding numerical data
- `price * quantity` is an **instruction** that multiplies the variables using the `*` operator
- `print()` is an **instruction** that displays the result
- `f""` is a special kind of text string that allows us to embed expressions and variables in strings using curly braces

**Summary:**

Every program you write will consist of:
1. **Creating variables** to store information
2. **Writing instructions** to process and use that information with operators
3. **Repeating** this cycle to build complex functionality

Master these concepts—variables, instructions, operators, and operands—and you'll be able to write programs in any programming language!

### Functions

Functions are one of the most powerful features in programming. They allow you to encapsulate code into reusable blocks, making your programs more organized, readable, and maintainable.

A function is defined by four key aspects:

* The **name** of the function: This is how we identify and call the function in our code.
* The **parameters** (also called arguments): These are the inputs that the function can receive. They are specified as a comma-separated list inside parentheses.
* The **return value**: This is the output that the function produces. In Python, functions can return any type of value (or nothing at all).
* The **function body**: This is the block of code that executes when the function is called.

In Python, functions are defined using the `def` keyword, followed by the function name, parameters in parentheses, and a colon. The function body must be indented.

Here's the basic syntax:

```python
def function_name(parameter1, parameter2, ...):
    # Function body
    # Code goes here
    return result  # Optional return statement
```

#### Example: Converting Meters to Centimeters

Let's create a function that converts meters to centimeters. This function will take a distance in meters as input and return the equivalent distance in centimeters.

```python
def meters_to_cm(meters):
    centimeters = meters * 100
    return centimeters
```

The `return` statement specifies what value the function should output when called.

#### Using the Function

Here's how we can use our `meters_to_cm` function:

```python
# Call the function and store the result
distance_m = 5.5
distance_cm = meters_to_cm(distance_m)

print(f"{distance_m} meters equals {distance_cm} centimeters")
```

Functions make code reusable and help break down complex problems into smaller, manageable parts. As you progress in programming, you'll learn about more advanced function concepts like default parameters, variable-length arguments, and lambda functions.
