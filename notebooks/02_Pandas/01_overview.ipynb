Pandas is the go-to library for data analysis in Python. Its name comes from “panel data” (an econometric term). It draws inspiration from R’s features but leverages the power of this general-purpose language.

Pandas includes all the features necessary for the data analysis process: loading, filtering, processing, summarizing, grouping, storing, and visualizing. In short, it’s like a spreadsheet —for example, Excel— but with much more potential.

All the work we’ll do is based on the basic data structure: the ``dataFrame``.

A ``dataFrame`` is a two-dimensional object that contains information. It can also be viewed as a spreadsheet, as a table in an entity-relationship model, or as a collection in a non-relational database.

__Filtering Data__

Filtering allows us to select rows that satisfy a logical condition.  For example, to find out which countries have an adult literacy rate of over 70%, we can do the following:

```python
literacy = df_who[df_who['Adult literacy rate (%)'] > 70][["Country","Adult literacy rate (%)"]]
literacy
```

This returns only rows where the condition is True and any selected columns. We can filter an specific value also:

```python
literacy[literacy["Country"] == "Italy"]
```

__How it works?__

We create a boolean condition:

```python
selection = df_who['Adult literacy rate (%)'] > 70
```

Then we apply the condition, only returning the rows where the condition is True:

```python
df_who[selection][["Country", "Adult literacy rate (%)"]]
```

We can filter multiple conditions using: 
- __&__ : AND
- __|__ : OR

Let's see the previous example, but filtering only the countries from Europe ( Continent value = 2)

```python
df_who[
    (df_who['Adult literacy rate (%)'] > 70) &
    (df_who[df_who.columns[2]] == 2)
]
```

<div class="alert alert-warning">
    <b> VIII. <i>No Numerical Series</i></b>
    <a id="nonumerical"></a></div>

When a column contains text (like country names), Pandas treats it as a Series of strings.

```python
df_who["Country"]
```

``.str``: To apply string methods to a column, we use .str. This gives access to vectorized string functions.

```python
df_who["Country"].str
```

``casefold()``: Converts all text to lowercase,

```python
df_who.Country.str.casefold()
```

``zfill(width)``:Pads each string with zeros on the left until it reaches the specified length. It is useful to formatting text to fixed width or IDs/Codes.

```python
df_who.Country.str.zfill(12)
```
